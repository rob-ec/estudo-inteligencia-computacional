## Natureza dos processos

Para conseguir implementar os Modelos de I.C. precisamos entender a natureza dos processos que estão sendo observados, de maneira a generalizar os tipos de processos podemos dividi-los em dois grupos, os Processo ou Sistemas Determinísticos e os Processos ou Sistemas Estocásticos. 

###### importações iniciais

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Determinístico

Em um sistema ou processo determinístico dado uma entrada X a saída sempre será o mesmo Y, a saída é exata e inalterada.

Todos os componentes da função são exatos.

**Ex.:**

$$y = w_0 + w_1 x$$

$$y = w_0 + w_1 x + w_2 x^2$$

$$y = w_0 + \cos(x) \cdot e^{-x}$$

#### Demonstração

```python
# Dados de entrada

x = np.array([value for value in np.arange(-2.0, 2.0, 0.1)])
```

$$w = [w_0, w_1]$$

$$y_i = w_0 + w_1 \cdot x_i, \forall i = 1...n$$

```python
w = [0.01, 0.3]

y = [w[0] + w[1] * xi for xi in x]

plt.plot(x,y)
```

$$w = [w_0, w_1, w_2]$$

$$y_i = w_0 + w_1 \cdot x_i + w_2 \cdot w_i^2, \forall i = 1...n$$

```python
w = [0.05, -0.1, 0.6, 0.3]

y = [w[0] + w[1] * xi + w[2] * xi**2 for xi in x]

plt.plot(x,y)
```

### Estocástico (Não Determinístico)

Em um sistema ou processo estocástico dado uma entrada X a saída nem sempre será o mesmo Y. Y segue uma distribuição de probabilidade $P(Y|X)$ e sempre há pelo menos um (1) componente aleatório na função.

**Ex.:**

$$y = w_0 + w_1 x + \mathcal{U}(\min = -1, \max = 1)$$

$$y = w_0 + w_1 x + \mathcal{N}(\min = 0, \max = 1)$$

#### O que gera a incerteza?

##### Incerteza intrínseca

A natureza do processo é probabilistica por definição.

##### Incerteza extrínseca

A incerteza é gerada por problemas de medição, precisão, armazenamento, arredondamento, etc.

#### Exemplos

- Processos econômicos e finânceiros
- Processos geológicos, meteorológicos, climáticos e hidrológicos
- Processos Sócio-Demográficos
- Processos Biológicos

$$\sim$$

#### Distribuição de Probabilidade

Uma distribuição de probabilidade é uma função que descreve a incerteza associada aos valores de uma variável aleatórea $X$.

$$X \sim P(parâmetros)$$

Ou seja, $X$ é distribuído de acordo com $P$ dados os $parâmetros$

##### Função de densidade de probabilidade (FDP)

A função de densidade de probabilidade (FDP) é uma função que associa à cada valor $x \in X$ um valor em $[0,1]$, tal que

$$P: X \rightarrow [0,1]$$

$$\sum_{x \in X} P(X) = 1$$

##### Exemplos de distribuições de probabilidade

|        Distribuição        |   Nome   |                           Densidade                            |
|----------------------------|----------|----------------------------------------------------------------|
| $\mathcal{U}(\min,\max)$   | Uniforme | $P(X) = {1 \over {(\max - \min)}}$                             |
| $\mathcal{N}(\mu, \sigma)$ | Normal   | $P(X) = {1 \over {\sigma \sqrt{2}\pi e^{-1/2(X-\mu)/\sigma}}}$ |

#### Demonstração

$$w = [w_0, w_1]$$

$$y_i = w_0 + w_1 \cdot x_i + \mathcal{U}(\min = -0.5, \max = 0.5)$$

$$y \sim \mathcal{U}(\min = -0.5, \max = 0.5)$$

```python
w = [0.01, 0.3]

y = [w[0] + w[1] * xi + np.random.uniform(-0.5, 0.5) for xi in x ]

plt.plot(x,y)
```

$$w = [w_0, w_1]$$

$$y_i = w_0 + w_1 \cdot x_i + \mathcal{N}(\mu = 0, \sigma = 0.1)$$

$$y \sim \mathcal{N}(\mu = 0, \sigma = 0.1)$$

```python
w = [0.01, 0.3]

y = [w[0] + w[1] * xi + np.random.normal(0, 0.1) for xi in x ]

plt.plot(x,y)
```

#### Distribuições de Probabilidade

$$x \sim \mathcal{U}(\min, \max) \implies P(X) = {{1}/{(\max - \min)}}$$

```python
# Cria um gráfico com largura 15, altura 5, com 1 linha e 2 colunas 
fig, ax = plt.subplots( nrows = 1, ncols = 2 , figsize = [15, 5] )

# Gera n valores da distribuição Uniforme com mínimo -0.5 e máximo 0.5
uniforme = np.random.uniform(-0.5, 0.5, 1000)

ax[0].plot(uniforme)
ax[0].set_title('Valores')
ax[1].hist(uniforme)
ax[1].set_title('Histograma')
```

$$x \sim \mathcal{N}(\mu, \sigma) \implies P(x) = {1 \over {\sigma \sqrt{2}\pi}} \cdot e^{{x-\mu}\over{2\sigma}}$$

```python
# Cria um gráfico com largura 15, altura 5, com 1 linha e 2 colunas 
fig, ax = plt.subplots( nrows = 1, ncols = 2 , figsize = [15, 5] )

# Gera n valores da distribuição Uniforme com mínimo -0.5 e máximo 0.5
normal = np.random.normal(0, 0.5, 1000)

ax[0].plot(normal)
ax[0].set_title('Valores')
ax[1].hist(normal)
ax[1].set_title('Histograma')
```
